```python
# Market Basket Optimization Associated Rule Learning - Apriori

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/shivang98/Market-Basket-Optimization/master/Market_Basket_Optimisation.csv', header = None)

#header = None coz there is no header in real dataset and if None is not givem, the first row will be considered as header
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.values.astype('str')   #we convert all values in df from object to str
```

```python
#we put all items in a list for each row i.e basket

transactions = df.tolist()
transactions

#OR
# transactions = []
# for i in range(0, 7501):
#     transactions.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
# Training Apriori on dataset

from apyori import apriori

rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)

#we are setting minimum values for support, conf, lift, and min length i.e combination
```

```python
# visualising results
results = list(rules)
results
```

```python
# RelationRecord( items                =  frozenset({'chicken', 'light cream'}),
#                 support              =  0.004532728969470737,
#                 ordered_statistics   =  [OrderedStatistic(items_base = frozenset({'light cream'}),
#                                        items_add                     = frozenset({'chicken'}), 
#                                        confidence                    = 0.29059829059829057,
#                                        lift                          = 4.84395061728395)]
#               )

for item in results:
    
#   first index of inner list contains base item and add item
    pair = item[0]

    combination = [x for x in pair]
    print('Rule       : ' , combination[0] , '->' , combination[1])
    
#   second index of the inner list
    print('Support    : ' , item[1])
    
#   third index of list located at 0th of the 3rd index of the inner list
    
    print('Confidence : ', item[2][0][2])
    print('Lift       : ', item[2][0][3])
    print('===========================================')
```
