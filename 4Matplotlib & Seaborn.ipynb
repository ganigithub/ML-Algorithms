### seaborn

```python
import seaborn as sns
sns.boxplot(x='Gender', y='Age', data=mydata)

# also called 5 number summary coz it gives Max, 75%, 50%, 25%, Min
# helps to identify ouliers. separate diamond shape points are outliers
# we can see there is outlier for Female which is more than Q3 + 1.5IQR hence will not be considered maxima.
# we can infer that males are fitness aware for longer period of time coz max age for males is more than female.
```

```python
sns.boxplot(x='Age', y='Gender', data=mydata)
```

```python
sns.boxplot(x='MaritalStatus', y='Age', data=mydata)
```

```python
pd.crosstab(mydata['Product'], mydata['Gender'])
```

```python
sns.countplot(x='Product', hue = 'Gender', data = mydata) #countplot is visual representation for crosstab
```

```python
sns.pairplot(mydata, diag_kind = 'kde')  #diag_kind is diagonal kind which is 'kde' (kernel density estimate)

#the diagonals should be straight lines but since we are using kde, it converts it to probability density.

#for eg. we can infer form age vs income that age & income has positive correlation upto some point.
```

```python
sns.pairplot(mydata)   #if no diag_kind provided, diagonal will be a histogram
```

```python
sns.distplot(mydata['Age'])

#distplot is distrubution plot ie variation in data distribution.
```
